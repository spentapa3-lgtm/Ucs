```python
import heapq

def uniform_cost_search(graph, start, goal):
    pq = [(0, start, [start])]
    explored = set()

    while pq:
        cost, node, path = heapq.heappop(pq)

        print("Node selected:", node)
        print("Cost so far:", cost)
        print("Path:", path)
        print("----------------------")

        if node == goal:
            print("Goal reached!")
            return cost, path

        if node in explored:
            continue

        explored.add(node)

        for neighbor, edge_cost in graph.get(node, []):
            if neighbor not in explored:
                heapq.heappush(pq, (cost + edge_cost, neighbor, path + [neighbor]))

    return float("inf"), []


graph = {
    'A': [('B', 2), ('C', 4)],
    'B': [('D', 3), ('E', 4)],
    'C': [('F', 5)],
    'D': [],
    'E': [('G', 3)],
    'F': [('H', 4), ('I', 6)],
    'G': [('J', 2)],
    'H': [],
    'I': [('K', 5)],
    'J': [('L', 2)],
    'K': [('M', 3)],
    'L': [],
    'M': [('N', 4)],
    'N': []
}

start = 'A'
goal = 'N'

cost, path = uniform_cost_search(graph, start, goal)

print("\nFinal Result:")
print("Optimal Cost:", cost)
print("Optimal Path:", " -> ".join(path))
```
